```python
df = df.head(100)

nh3_list = df['nh3'].tolist()
light_list = df['light'].tolist()
size=df['nh3']
size2=df['light']

trace1 = go.Scatter(x=df['dateTime'],
                    y=nh3_list,
                    name='nh3',
                    mode='markers',
                    marker=dict(
                        size=size,
                        sizemode='area',
                        sizeref=2.*max(size)/(40.**2),
                        sizemin=4,
                        color='pink',
                        line=dict(width=1, color="#CC0066")
                    ), 
                    line=dict(width=1.5))

trace2 = go.Scatter(x = df['dateTime'],
                    y = light_list,
                    name='humidity',
                    mode='markers',
                    marker=dict(
                        size=size2,
                        sizemode='area',
                        sizeref=2.*max(size)/(40.**2),
                        sizemin=4,
                        color='aliceblue',
                        line=dict(width=1, color='grey')
                    ), 
                    line=dict(width=1.5))

frames = [dict(data= [dict(type='scatter',
                           x=df['dateTime'][:k+1],
                           y=nh3_list[:k+1]),
                      dict(type='scatter',
                           x=df['dateTime'][:k+1],
                           y=light_list[:k+1])],
               traces= [0, 1],  #this means that  frames[k]['data'][0]  updates trace1, and   frames[k]['data'][1], trace2 
              )for k  in  range(1, len(nh3_list)-1)] 

layout = go.Layout(showlegend=True,
                   hovermode='closest', 
                   updatemenus=[dict(type='buttons', showactive=False,
                                y=1.05,
                                x=1.15,
                                xanchor='left',
                                yanchor='top',
                                pad=dict(t=0, r=10),
                                buttons=[dict(label='Play',
                                              method='animate',
                                              args=[None, 
                                                    dict(frame=dict(duration=4, 
                                                                    redraw=False),
                                                         transition=dict(duration=0),
                                                         fromcurrent=True,
                                                         mode='immediate')])])])

fig = go.Figure(data=[trace1, trace2], frames=frames, layout=layout)
fig.show()
```
