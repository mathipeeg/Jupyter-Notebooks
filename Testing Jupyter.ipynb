## Simple plotting example

This is a plot of the $sinc$ function

$$
f(x) = \frac{\sin(x)}{x}
$$

```python
%matplotlib inline
%config InlineBackend.figure_format = 'svg'

pi = np.pi
x = np.linspace(-4*pi, 4*pi, 1000) 
plt.rcParams['figure.figsize'] = (11, 4)
plt.rcParams.update({'font.size': 10})
plt.yticks([0,1], ['$0$', '$1$'])
plt.xticks([-4*pi, -2*pi, 2*pi, 4*pi],
          ['$-4\pi$', '$-2\pi$', '$0$', '$2\pi$', '$4\pi$'])
plt.axhline(0, color='black', lw=1)
plt.plot(x, np.sin(x)/x, label=r'$f(x) = \frac{\sin(x)}{x}$')
plt.ylim(-0.5, 1.5)
plt.title('The $sinc$ function')
plt.legend(loc='best', fontsize=15)
plt.show()
```

## Animated Plotting

```python
import numpy as np
import matplotlib.pyplot as plt

def rect(x, B):
    """
    create a rectangle function
    returns a numpy array that is 1 if |x| < w and 0 if |x| > w
    B is the rectangle width centered at 0
    x is the number of points in the array
    """
    
    B = int(B)
    x = int(x)
    
    high = np.ones(B)
    low1 = np.zeros(int(x/2 - B/2))    
    x1 = np.append(low1, high)
    rect = np.append(x1, low1)
    
    if x > len(rect):
        rect = np.append(rect, 0)
    elif x < len(rect):
        rect = rect[:-1]

    return rect    
```

```python
%matplotlib notebook
from matplotlib.animation import FuncAnimation

# constants and x array
pi = np.pi
length = 2000
x = np.linspace(-1, 1, length)

fig,(ax1, ax2) = plt.subplots(2, figsize=(12, 6))
sinc, = ax1.plot(x, np.sin(x), '-b')
box, = ax2.plot(x, np.sin(x), '-r')

def animate(B):
    """
    this function gets called by FuncAnimation
    each time called, it will replot with a different width "B"
    
    B: rect width
    
    return:
        sinc: ydata
        box: ydata
    """
    
    f = rect(len(x), B)
    box.set_ydata(f)
    
    F = (B / length) * np.sin(x * B/ 2) / (x * B / 2)
    sinc.set_ydata(F)
    
    ax1.set_ylim(np.min(F), np.max(F)) # Dette gør, at boksen som den øverste graf er i, altid har højde af det grafens højde er.
    plt.setp(ax1, xticks=[-0.25, 0.25], xticklabels=['-1/4', '1/4'],
            yticks=[0, np.max(F)], yticklabels=['0', 'B={:.2f}'.format((B/length))]) #format B / length, float og to decimaltal
    
    plt.setp(ax2, yticks=[0, 1],
            xticks=[-1, -1 * B / length, 1 *B / length, 1], xticklabels=['-1', '-B/2', 'B/2', '1'])
    
def init():
    """
    initialise the figure
    """
    
    ax2.set_ylim(-0.2, 1.1)
    ax1.set_xlim(-0.25, 0.25)
    ax2.set_xlim(-1, 1)
    ax1.axhline(0, color='black', lw=1)
    ax2.axhline(0, color='black', lw=1)
    plt.rcParams.update({'font.size': 14})
    
    return sinc, box,

step = 10
steps = np.append(np.arange(10, 1000, step), np.arange(1000, 10, -1 * step))
ani = FuncAnimation(fig, animate, steps, init_func=init, interval=50, blit=True)
plt.show()
```

## Interactive Widgets

```python
import ipywidgets as widgets
from IPython.display import display
# plt and np

%matplotlib notebook
```

$A=amplitude$

$\phi=  phase$

$\nu= frequency$

$A \sin(2 \pi(\nu x + \phi))$

```python
x = np.linspace(0, 2, 100)
fig, ax = plt.subplots(1, figsize=(10, 4))
plt.suptitle('Sine Wave :D woo')

def update_plot(amp, phase, freq):
    """
    this function is linked to the sliders and 
    it replots the sine wave when the sliders are changed
    """
    ax.clear()
    
    units = 'amp = {} $(psi)$ \nphase = {} $(s)$ \nfreq = {} $(Hz)$'
    y = amp * np.sin(2 * np.pi * (freq * x + phase))
    
    ax.plot(x, y, label=units.format(amp, phase, freq))
    ax.legend(loc=1)
    ax.set_xlabel('$(s)$')
    ax.set_ylabel('$(psi)$')
    ax.set_xlim(x[0], x[-1]) #intet hvidt space i siderne
    plt.show()

amp = widgets.FloatSlider(min=1, max=10, value=5, description='Amp:')
phase = widgets.FloatSlider(min=0, max=5, value=0, description='Phase:')
freq = widgets.FloatSlider(min=1, max=10, value=1, description='Frequency:')

widgets.interactive(update_plot, amp=amp, phase=phase, freq=freq)
```
